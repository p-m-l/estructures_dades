```python
# Exercici 1

from sklearn.linear_model import LinearRegression

# Carrega el dataset en un DataFrame de Pandas
df = pd.read_csv("Lligues europees UEFA.csv",  sep=';', encoding='latin')

# Selecciona les columnes "Pts" i "GF" del DataFrame
X = df[["Pts"]]
y = df[["GF"]]

# Crea un objecte de regressió linear
model = LinearRegression()

# Entrena el model amb les dades seleccionades
model.fit(X, y)

# Mostra la relació entre "Pts" i "GF" segons el model
print(model.coef_)

```

```python
from scipy import stats

# Seleccionem les columnes "Pts" i "GF" del DataFrame
X = df["Pts"]
y = df["GF"]

# Calculem el test de Student i el p-valor
t, p_value = stats.ttest_ind(X, y)

# Si el p-valor és menor que l'alfa de 5% (0.05), rebutja la hipòtesi nul·la
if p_value < 0.05:
    print("Es pot rebutjar la hipòtesi nul·la amb un alfa de 5%.")
else:
    print("No es pot rebutjar la hipòtesi nul·la amb un alfa de 5%.")
print("P-Value is: ",p_value)
```

```python
# Exercici 2

from scipy import stats

# Seleccionem les columnes "W" i "MP" del DataFrame
x1 = df["W"]
y1 = df["Pts/G"]

# Calculem el test de Student i el p-valor
t1, p_value1 = stats.ttest_ind(x1, y1)

# Si el p-valor és menor que l'alfa de 5% (0.05), rebutja la hipòtesi nul·la
if p_value1 < 0.05:
    print("Es pot rebutjar la hipòtesi nul·la amb un alfa de 5%.")
else:
    print("No es pot rebutjar la hipòtesi nul·la amb un alfa de 5%.")

print("P-Value is: ",p_value1, "≈ 0. Per tant veiem que tenen una forta relació entre sí")
```

```python
# Exercici 3

from scipy import stats

# Seleccionem les columnes "W", "D" i "L" del DataFrame
X1 = df["W"]
X2 = df["D"]
X3 = df["L"]

# Crea un objecte Anova amb les dades seleccionades
anova = stats.f_oneway(X1, X2, X3)

# Si el p-valor és menor que l'alfa de 5% (0.05), rebutja la hipòtesi nul·la
# Si el p-valor és menor que l'alfa de 5% (0.05), rebutja la hipòtesi nul·la
if anova.pvalue < 0.05:
    print("Es pot rebutjar la hipòtesi nul·la amb un alfa de 5%.")
else:
    print("No es pot rebutjar la hipòtesi nul·la amb un alfa de 5%.")
print("P-Value is: ", anova.pvalue, "≈ 0. Per tant veiem que tenen una forta relació entre sí")
```
